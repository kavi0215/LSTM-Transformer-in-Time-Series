```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv(r"C:\Users\anjal\Downloads\archive (1)\stock_market_data\nasdaq\csv\AAPL.csv")
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df = df.sort_values('Date')
print(df.head())
```

```python
features = ['Open', 'High', 'Low', 'Close', 'Volume']
target_col = 'Close'
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[features])
scaled_df = pd.DataFrame(scaled_data, columns=features)
```

```python
SEQ_LEN = 30

def create_sequences(data, target_index):
    X, y = [], []
    for i in range(len(data) - SEQ_LEN):
        X.append(data[i:i+SEQ_LEN])
        y.append(data[i+SEQ_LEN, target_index])
    return np.array(X), np.array(y)
X, y = create_sequences(scaled_df.values, features.index(target_col))
print(X.shape, y.shape)
```

```python
split = int(0.8 * len(X))

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)

X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)
```

```python
class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_loader = DataLoader(
    TimeSeriesDataset(X_train, y_train),
    batch_size=32,
    shuffle=False
)

test_loader = DataLoader(
    TimeSeriesDataset(X_test, y_test),
    batch_size=32,
    shuffle=False
)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size, hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=0.2
        )
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        return self.fc(out).squeeze()
```

```python
model = LSTMModel(input_size=len(features))
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
EPOCHS = 10   # reduced epochs (your earlier request)

for epoch in range(EPOCHS):
    model.train()
    epoch_loss = 0

    for xb, yb in train_loader:
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {epoch_loss/len(train_loader):.6f}")
```

```python
model.eval()
preds = []

with torch.no_grad():
    for xb, _ in test_loader:
        preds.extend(model(xb).numpy())

preds = np.array(preds)

# Inverse scaling
close_index = features.index('Close')
close_min = scaler.data_min_[close_index]
close_max = scaler.data_max_[close_index]

preds_inv = preds * (close_max - close_min) + close_min
y_test_inv = y_test.numpy() * (close_max - close_min) + close_min

mae = mean_absolute_error(y_test_inv, preds_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, preds_inv))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
plt.figure(figsize=(12,5))
plt.plot(y_test_inv[:200], label='Actual')
plt.plot(preds_inv[:200], label='Predicted')
plt.legend()
plt.title("LSTM Forecast vs Actual")
plt.show()
```

```python
window = 500
step = 100
wf_errors = []

for start in range(0, len(X) - window - step, step):
    end = start + window
    test_end = end + step

    X_tr = torch.tensor(X[start:end], dtype=torch.float32)
    y_tr = torch.tensor(y[start:end], dtype=torch.float32)

    X_te = torch.tensor(X[end:test_end], dtype=torch.float32)
    y_te = y[end:test_end]

    model = LSTMModel(len(features))
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    for _ in range(5):
        preds = model(X_tr)
        loss = criterion(preds, y_tr)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    with torch.no_grad():
        p = model(X_te).numpy()

    p = p * (close_max - close_min) + close_min
    y_true = y_te * (close_max - close_min) + close_min

    wf_errors.append(mean_absolute_error(y_true, p))

print("Walk-forward MAE:", np.mean(wf_errors))
```

```python
class TimeSeriesTransformer(nn.Module):
    def __init__(
        self,
        input_dim,
        d_model=64,
        nhead=4,
        num_layers=2,
        dim_feedforward=128,
        dropout=0.1
    ):
        super().__init__()

        self.input_projection = nn.Linear(input_dim, d_model)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            dim_feedforward=dim_feedforward,
            dropout=dropout,
            batch_first=True
        )

        self.transformer_encoder = nn.TransformerEncoder(
            encoder_layer,
            num_layers=num_layers
        )

        self.fc = nn.Linear(d_model, 1)

    def forward(self, x):
        # x shape: (batch, seq_len, input_dim)
        x = self.input_projection(x)
        x = self.transformer_encoder(x)
        x = x[:, -1, :]     # last timestep
        return self.fc(x).squeeze()
```

```python
transformer_model = TimeSeriesTransformer(
    input_dim=len(features),
    d_model=64,
    nhead=4,
    num_layers=2
)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(transformer_model.parameters(), lr=0.001)
```

```python
EPOCHS = 10

for epoch in range(EPOCHS):
    transformer_model.train()
    epoch_loss = 0

    for xb, yb in train_loader:
        optimizer.zero_grad()
        preds = transformer_model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {epoch_loss/len(train_loader):.6f}")
```

```python
transformer_model.eval()
preds = []

with torch.no_grad():
    for xb, _ in test_loader:
        preds.extend(transformer_model(xb).numpy())

preds = np.array(preds)

# Inverse scaling
close_index = features.index('Close')
close_min = scaler.data_min_[close_index]
close_max = scaler.data_max_[close_index]

preds_inv = preds * (close_max - close_min) + close_min
y_test_inv = y_test.numpy() * (close_max - close_min) + close_min

mae_t = mean_absolute_error(y_test_inv, preds_inv)
rmse_t = np.sqrt(mean_squared_error(y_test_inv, preds_inv))

print("Transformer MAE:", mae_t)
print("Transformer RMSE:", rmse_t)
```

```python
plt.figure(figsize=(12,5))
plt.plot(y_test_inv[:200], label='Actual')
plt.plot(preds_inv[:200], label='Transformer Prediction')
plt.legend()
plt.title("Transformer Forecast vs Actual")
plt.show()
```
